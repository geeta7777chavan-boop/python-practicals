**1. What is the importance of data visualization in data analysis? How does visualization help in exploratory data analysis (EDA)?**

Data Visualization is the graphical representation of information and data using charts, graphs, plots, and maps. It plays a vital role in data analysis by converting complex numerical data into an easily understandable visual format.

**Importance of Data Visualization:**


*   Makes it easier to identify patterns, trends, correlations, and outliers in the data.

*   Helps in communicating insights effectively to technical and non-technical stakeholders.
*   Reduces the time required to understand large volumes of data.


*   Supports better decision-making by revealing hidden information that raw numbers cannot show.

**Role in Exploratory Data Analysis (EDA):**

Visualization is one of the most important steps in EDA. It helps analysts quickly understand the distribution of data, relationships between variables, presence of missing values or anomalies, and overall data quality before applying statistical models or machine learning algorithms.

**3. What is the purpose of the pyplot module in Matplotlib?**

matplotlib.pyplot is a state-based interface of Matplotlib that provides a MATLAB-like plotting environment. It is the most commonly used module for creating visualizations in Matplotlib.

**Purpose of pyplot :**

(a) Simplifies the creation of figures and axes.

(b) Provides simple functions to add titles, labels, legends, grids, etc.

(c) Maintains the current figure and axes internally (stateful interface).

(d) Allows quick plotting with minimal code (e.g., plt.plot(), plt.title(), plt.show()).

It acts as the main entry point for most users when working with Matplotlib.

**4. Explain the process of creating a basic line plot.**

The process of creating a basic line plot in Matplotlib involves the following steps:

1. Import the required libraries (import matplotlib.pyplot as plt).
2. Prepare the data (x-axis and y-axis values).
3. Use plt.plot(x, y) to create the line plot.
4. Add enhancements such as title (plt.title()), x-label (plt.xlabel()), y-label (plt.ylabel()), and legend.
5. Display the plot using plt.show().

This simple workflow allows quick visualization of trends and relationships between variables.

**5. Why is customization important in data visualization? What customization options are available in Matplotlib?**

Customization is important because default plots are often plain and may not effectively communicate the intended message. Well-customized plots improve readability, aesthetics, and clarity, making insights more impactful.

**Customization options available in Matplotlib:**

i. Changing figure size `(plt.figure(figsize=(width, height)))`

ii. Setting titles, axis labels, and font sizes

iii. Adding grid lines `(plt.grid())`

iv. Customizing colors, line styles, and markers

v. Adding legends and annotations

vi. Changing plot style using `plt.style.use()`

vii. Rotating tick labels and adjusting axis limits

viii. Adding subplots for multiple plots in one figure

These options allow users to create publication-quality and presentation-ready visualizations.

# **Program :**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
plt.style.use('seaborn-v0_8')
sns.set(style="whitegrid")

# Creating a sample Student Performance Dataset
data = {
    'Student_ID': range(1, 21),
    'Math': [85, 78, 92, 65, 88, 72, 95, 68, 82, 90, 75, 87, 80, 70, 93, 77, 84, 66, 89, 91],
    'Science': [80, 85, 88, 70, 82, 75, 90, 72, 85, 87, 78, 83, 79, 74, 91, 76, 86, 69, 88, 92],
    'English': [75, 82, 90, 68, 85, 80, 88, 65, 78, 92, 70, 84, 81, 73, 89, 79, 83, 67, 87, 85],
    'History': [70, 75, 80, 60, 78, 72, 85, 62, 76, 82, 68, 79, 74, 71, 88, 73, 77, 65, 81, 84]
}

df = pd.DataFrame(data)

# Display basic information
print("Dataset Preview:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
```

**1. Bar Plot – Average Marks Comparison Across Subjects**

```python
# Calculate average marks per subject
avg_marks = df[['Math', 'Science', 'English', 'History']].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_marks.index, y=avg_marks.values, palette='viridis')
plt.title('Average Marks Comparison Across Subjects', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.ylim(0, 100)
plt.show()
```

**2. Histogram – Distribution of Marks in Mathematics**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Math'], bins=10, kde=True, color='blue')
plt.title('Distribution of Mathematics Marks', fontsize=14)
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.axvline(df['Math'].mean(), color='red', linestyle='--', label=f'Mean: {df["Math"].mean():.1f}')
plt.legend()
plt.show()
```

**3. Box Plot – Performance Comparison Across All Subjects**

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=df[['Math', 'Science', 'English', 'History']], palette='Set2')
plt.title('Box Plot: Student Performance Across Subjects', fontsize=14)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

**4. Line Plot – Individual Student Performance Trend**

```python
# Select first 8 students for clarity
student_data = df.head(8).set_index('Student_ID')

plt.figure(figsize=(12, 6))
for subject in ['Math', 'Science', 'English', 'History']:
    plt.plot(student_data.index, student_data[subject], marker='o', label=subject)

plt.title('Performance Trend of Students Across Subjects', fontsize=14)
plt.xlabel('Student ID')
plt.ylabel('Marks')
plt.legend(title='Subjects')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
