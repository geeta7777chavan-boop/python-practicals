# **Theory:**

Sequence data types in Python are collections of items that are stored in a specific order. These data types allow indexing, slicing, and iteration, making them essential for handling and manipulating data efficiently.


**1. Types of Sequence Data Types**


**a. Strings**

Strings are immutable sequences of characters enclosed in single, double, or triple quotes.

Example: "Hello"

Support indexing and slicing
Immutable (cannot be changed after creation)

**b. Lists**

Lists are ordered, mutable collections of elements.

Example: [1, 2, 3, "apple"]

Allow duplicate values
Can store different data types
Support dynamic modification (add/remove elements)

**c. Arrays**

Arrays are similar to lists but store elements of the same data type using the array module.

Example: array('i', [1, 2, 3])

More memory efficient than lists
Used for numerical data

**d. Tuples**

Tuples are ordered, immutable collections.

Example: (1, 2, 3)

Faster than lists
Cannot be modified after creation

**e. Dictionary**

Dictionaries store data in key-value pairs.

Example: {"name": "John", "age": 20}

Keys are unique
Unordered (but maintain insertion order in modern Python)

**f. Sets**

Sets are unordered collections of unique elements.

Example: {1, 2, 3}

Automatically remove duplicates
Useful for mathematical operations like union and intersection

**g. Range**

Range represents a sequence of numbers.

Example: range(1, 10)

Commonly used in loops

Memory efficient (does not store all values)

**2. Indexing and Slicing**

**Indexing**: Accessing elements using position



Example: list[0] gives first element

**Negative indexing:** Access from end

Example: list[-1] gives last element

**Slicing:** Extracting a subset

Syntax: sequence[start:end:step]

Example: list[1:4]

Indexing and slicing work similarly in strings, lists, tuples, and ranges, but not in dictionaries and sets since they are not position-based.

**3. Difference Between List and Tuple**

| Feature     | List    | Tuple     |
| ----------- | ------- | --------- |
| Mutability  | Mutable | Immutable |
| Performance | Slower  | Faster    |
| Syntax      | []      | ()        |





  **Why prefer tuple?**
*   Used when data should not change (e.g., coordinates)
*   Faster and memory efficient

*   Safe from accidental modification

**4. Arrays vs Lists**
| Feature   | List   | Array                  |
| --------- | ------ | ---------------------- |
| Data Type | Mixed  | Same type only         |
| Memory    | More   | Less                   |
| Speed     | Slower | Faster for numeric ops |


**Use array when:**


*   Working with large numerical datasets
*   Memory efficiency is important



**5. Dictionaries (Internal Working)**

Dictionaries use a hash table internally:

Each key is hashed to a unique index
Values are stored at computed positions

**Why faster?**

Lookup time is O(1) (constant time)

No need to scan entire data like lists

**6. Sets vs Lists**
| Feature     | List          | Set           |
| ----------- | ------------- | ------------- |
| Order       | Ordered       | Unordered     |
| Duplicates  | Allowed       | Not allowed   |
| Performance | Slower search | Faster search |


**Handling duplicates:**

Sets automatically remove duplicate values

**Practical uses of sets:**

1. Removing duplicates

2. Membership testing (in)

3. Mathematical operations (union, intersection)

## **a) Strings**

# **b) Lists**

```python
# 2. Create a list of 5 integers and print max and min
lst = [10, 25, 3, 78, 45]

print("List:", lst)
print("Maximum:", max(lst))
print("Minimum:", min(lst))
```

```python
# 7. Insertion and deletion operations
lst = [1, 2, 3, 4]

# Insert
lst.append(5)
lst.insert(2, 10)

print("After insertion:", lst)

# Delete
lst.remove(3)
lst.pop()

print("After deletion:", lst)
```

# **c**) **Tuples**

```python
# 3. Tuple of student names
students = ("Amit", "Neha", "Ravi", "Sneha")

for name in students:
    print(name)
```

```python
# 8. Convert list to tuple and tuple to list
lst = [1, 2, 3]
tpl = tuple(lst)

print("Tuple:", tpl)

new_list = list(tpl)
print("List:", new_list)
```

# **d) Dictionary**

```python
# 4. Dictionary with roll number and name
student = {1: "Amit", 2: "Neha", 3: "Ravi"}

print("Keys:")
for key in student.keys():
    print(key)
```

```python
# 11. Word frequency
text = "python is easy and python is powerful"

words = text.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print("Word Frequency:", freq)
```

```python
# 13. Nested dictionary
students = {
    1: {"name": "Amit", "marks": 85},
    2: {"name": "Neha", "marks": 90}
}

print("Student 1 Name:", students[1]["name"])
```

# **e) Sets**

```python
# 5. Remove duplicates using set
lst = [1, 2, 2, 3, 4, 4, 5]

unique_set = set(lst)
print("Unique values:", unique_set)
```

```python
# 9. Set operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference:", set1 - set2)
```

# g) **Range**

```python
# 10. Generate range and convert to list
r = range(1, 11)

print("Range as list:", list(r))
```

```python
# 14. Compare performance of list vs tuple
import time

lst = list(range(1000000))
tpl = tuple(lst)

start = time.time()
for i in lst:
    pass
print("List time:", time.time() - start)

start = time.time()
for i in tpl:
    pass
print("Tuple time:", time.time() - start)
```
