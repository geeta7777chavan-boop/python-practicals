# **Theory**

**Introduction**

Correlation and regression are fundamental statistical techniques used to analyze relationships between variables. Correlation measures the strength and direction of a relationship, while regression helps in predicting one variable based on another. In this assignment, we study Karl Pearson’s correlation coefficient and simple linear regression, along with their practical implementation in Python.

**1. What is Correlation? How is it different from Regression?**


Correlation

Correlation measures the strength and direction of a relationship between two variables.

Value ranges from -1 to +1

+1 → Perfect positive relationship

-1 → Perfect negative relationship

0 → No relationship

Regression

Regression is used to predict the value of one variable based on another.

| Feature   | Correlation          | Regression                     |
| --------- | -------------------- | ------------------------------ |
| Purpose   | Measure relationship | Predict values                 |
| Output    | Single value (r)     | Equation (y = mx + c)          |
| Direction | Symmetric            | One dependent, one independent |
| Usage     | Analysis             | Prediction                     |


**2. What is Simple Linear Regression?**

Simple Linear Regression models the relationship between two variables using a straight line.

**y = mx + c**

Where:

y = dependent variable

x = independent variable

m = slope

c = intercept

Example

Predicting marks based on study hours:

More study hours → higher marks
The model finds the best-fit line

**3. How do you calculate Pearson correlation in Python?**

Formula :

$$
r = \frac{\sum (x - \bar{x})(y - \bar{y})}
{\sqrt{\sum (x - \bar{x})^2 \sum (y - \bar{y})^2}}
$$

Python Code :



```python
import numpy as np

# Sample data
hours = [1, 2, 3, 4, 5]
marks = [2, 4, 5, 4, 5]

# Correlation
correlation = np.corrcoef(hours, marks)[0,1]
print("Pearson Correlation:", correlation)
```

**4. What is Multicollinearity? Does it affect Simple Regression?**

Definition

Multicollinearity occurs when independent variables are highly correlated with each other.

Impact

Affects multiple regression models
Causes instability in coefficient estimation

Important Point

❌ Does NOT affect simple linear regression

✅ Because only one independent variable is used

**5. Strong Correlation but Poor Prediction Accuracy – Why?**

Even if correlation is high, prediction can fail due to:

Non-linear relationship (correlation only captures linear)

Outliers affecting model

Overfitting or underfitting

Small dataset

Hidden variables (missing features)

👉 Example:

Study hours vs marks may have high correlation, but stress, sleep, or difficulty level also affect marks.

# **Implementation**

```python
import numpy as np
import matplotlib.pyplot as plt

# Simple dataset
hours = np.array([1, 2, 3, 4, 5])
marks = np.array([2, 3, 4, 5, 6])

# 1. Pearson Correlation
r = np.corrcoef(hours, marks)[0,1]
print("Pearson Correlation (r):", r)

# 2. Linear Regression (y = mx + c)
m, c = np.polyfit(hours, marks, 1)
print("Slope (m):", m)
print("Intercept (c):", c)

# Predicted line
line = m * hours + c

# Plot
plt.figure()
plt.scatter(hours, marks, label="Data Points")
plt.plot(hours, line, label="Regression Line")

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()

plt.show()
```
