# **Theory**

**1. What is web scraping? Explain the role of the Requests library in web scraping.**

Web Scraping is the process of automatically extracting large amounts of data from websites. Instead of manually copying information, web scraping uses scripts to fetch web pages and pull out specific data such as text, images, links, prices, or tables.

**Role of the Requests Library:**

The requests library is used for making HTTP requests to websites. It acts as the first step in web scraping by:

1. Sending a GET request to the target URL.
2. Fetching the raw HTML content of the webpage.
3. Handling headers, cookies, and sessions if needed.

Without the requests library, we cannot retrieve the HTML content required for scraping. It is simple, clean, and widely used for downloading web pages in Python.

**2. What is BeautifulSoup? How does it help in parsing HTML?**

BeautifulSoup is a popular Python library used for parsing and extracting data from HTML and XML documents. It creates a parse tree from the raw HTML, making it easy to navigate, search, and modify the document.

**How it helps in parsing HTML:**

1. It converts messy, unstructured HTML into a well-organized parse tree.
2. Provides simple methods to find and extract specific tags, classes, or IDs.
3. Handles poorly formatted HTML gracefully.
4. Works well with different parsers like html.parser, lxml, or html5lib.

In short, while requests fetches the HTML, BeautifulSoup helps in understanding and extracting meaningful information from it.

**3. Differentiate between web scraping and web crawling.**

| Feature            | Web Scraping                                      | Web Crawling                                      |
|-------------------|--------------------------------------------------|--------------------------------------------------|
| Definition        | Extracting specific data from websites           | Automatically browsing and indexing web pages     |
| Purpose           | Data collection (targeted information)           | Discovering and indexing URLs                    |
| Focus             | Specific content (e.g., prices, text, images)    | Entire website or multiple pages                 |
| Scope             | Narrow and targeted                              | Broad and large-scale                           |
| Output            | Structured data (CSV, JSON, database)            | List of URLs or indexed pages                   |
| Tools Used        | BeautifulSoup, Scrapy                            | Scrapy, Selenium, search engine bots            |
| Example           | Extracting product prices from a site            | Google bot scanning websites                    |
| Usage             | Data analysis, research                          | Search engines, website indexing                |

Web scraping focuses on extracting specific data from a webpage, while web crawling involves navigating through multiple web pages to discover and index content.

**4. What are the commonly used methods in BeautifulSoup?**

BeautifulSoup provides several powerful methods to extract data:

`find()` – Returns the first matching tag.

`find_all()` – Returns a list of all matching tags.

`get_text()` – Extracts only the text content from a tag.

`get()` – Retrieves the value of an attribute (e.g., href, src, class).

`select()` – Uses CSS selectors to find elements (very powerful).

`select_one()` – Returns the first match using CSS selector.

`prettify()` – Returns a nicely formatted version of the HTML.

These methods make it easy to navigate through HTML tags like `<div>, <a>, <table>, <h1>`, etc.

**5. What are the ethical and legal considerations in web scraping?**

Web scraping must be done responsibly. Important considerations include:

**1) Legal Aspects:**

Respect the website’s robots.txt file, which tells which parts can or cannot be scraped.
Do not scrape websites that explicitly prohibit scraping in their Terms of Service.
Avoid scraping personal data without consent (may violate privacy laws like GDPR).

**2) Ethical Aspects:**

Do not overload the server with too many rapid requests (use time.sleep() to add delays).
Identify your scraper with a proper User-Agent header.
Do not scrape copyrighted content without permission.
Use scraping only for public, non-sensitive data and for legitimate purposes (research, analysis, education).


**Best Practice:** Always check the website’s terms, limit request rate, and consider using official APIs if available instead of scraping.

# **Program :**

```python
# Scenario: Collecting Disease Statistics from a Public Health Website

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import warnings
warnings.filterwarnings("ignore")

print("=== Web Scraping Disease Statistics ===\n")

# ==================== METHOD 1: Using Sample HTML (Safe for Colab) ====================

print("Using Sample Public Health Data (Simulated Table)...\n")

# Simulated HTML from a public health website (realistic structure)
html_content = """
<!DOCTYPE html>
<html>
<head><title>Public Health Disease Statistics 2025</title></head>
<body>
    <h1>Reported Infectious Disease Cases - India</h1>
    <table id="disease-table" border="1">
        <thead>
            <tr>
                <th>Disease Name</th>
                <th>Reported Cases</th>
                <th>Region</th>
                <th>Year</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Dengue</td>
                <td>245670</td>
                <td>Maharashtra</td>
                <td>2025</td>
            </tr>
            <tr>
                <td>Malaria</td>
                <td>189450</td>
                <td>Odisha</td>
                <td>2025</td>
            </tr>
            <tr>
                <td>Chikungunya</td>
                <td>67890</td>
                <td>Karnataka</td>
                <td>2025</td>
            </tr>
            <tr>
                <td>Tuberculosis</td>
                <td>1240000</td>
                <td>Uttar Pradesh</td>
                <td>2025</td>
            </tr>
            <tr>
                <td>COVID-19</td>
                <td>45230</td>
                <td>Delhi</td>
                <td>2025</td>
            </tr>
            <tr>
                <td>Typhoid</td>
                <td>98760</td>
                <td>Bihar</td>
                <td>2025</td>
            </tr>
        </tbody>
    </table>
</body>
</html>
"""

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table
table = soup.find('table', id='disease-table')

# Extract headers
headers = [th.text.strip() for th in table.find('thead').find_all('th')]
print("Extracted Headers:", headers)

# Extract rows
data = []
rows = table.find('tbody').find_all('tr')

for row in rows:
    cols = row.find_all('td')
    if len(cols) == 4:
        disease = cols[0].text.strip()
        cases = cols[1].text.strip().replace(',', '')   # remove commas
        region = cols[2].text.strip()
        year = cols[3].text.strip()

        data.append({
            'Disease_Name': disease,
            'Reported_Cases': int(cases),
            'Region': region,
            'Year': year
        })

# Convert to Pandas DataFrame
df = pd.DataFrame(data)

print("\nScraped Disease Statistics:")
print(df)

# Save to CSV for further analysis
df.to_csv('disease_statistics.csv', index=False)
print("\nData successfully saved to 'disease_statistics.csv'")

# ==================== Basic Analysis ====================

print("\n=== Quick Analysis ===")
print(f"Total Diseases Scraped: {len(df)}")
print(f"Total Reported Cases: {df['Reported_Cases'].sum():,}")
print("\nTop 3 Regions by Cases:")
print(df.groupby('Region')['Reported_Cases'].sum().sort_values(ascending=False).head(3))
```
