**1. What is data wrangling? Explain the difference between data wrangling and data preprocessing.**

Data Wrangling (also known as Data Munging) is the process of cleaning, transforming, structuring, and enriching raw data into a more suitable and usable format for analysis, visualization, and machine learning.

It includes tasks such as:


*   Handling missing values
*   Removing duplicates


*   Renaming columns
*   Changing data types


*   Merging multiple datasets
*   Creating new derived columns


*   Filtering and reshaping data









**Difference between Data Wrangling and Data Preprocessing:**

| Aspect        | Data Wrangling                                      | Data Preprocessing                                      |
|--------------|-----------------------------------------------------|---------------------------------------------------------|
| Definition    | Transforming raw data into a usable format          | Preparing data specifically for modeling                |
| Scope         | Broader (includes ingestion & transformation)       | Narrower (focuses on final model-ready data)            |
| Stage         | Early to middle stage                               | Later stage (after wrangling)                           |
| Main Goal     | Make data clean and structured                      | Optimize data for algorithms                            |
| Examples      | Merging files, pivoting, reshaping                  | Scaling, encoding categorical variables, train-test split |

In simple terms:

Data Wrangling → Makes messy data usable.

Data Preprocessing → Makes usable data suitable for machine learning algorithms.

**2. What is the importance of Exploratory Data Analysis (EDA) in Pandas?**

Exploratory Data Analysis (EDA) is the process of analyzing and summarizing the main characteristics of a dataset, often using visual and statistical methods, before applying formal modeling or machine learning techniques.

**Importance of EDA in Pandas:**


*   Helps understand the structure, distribution, and patterns in the data.

*   Identifies missing values, outliers, and anomalies.

*   Reveals relationships and correlations between variables.
*   Detects data quality issues early.


*   Guides feature selection and engineering decisions.


*  Prevents wrong assumptions during modeling.


*   Provides insights that drive business or research decisions.

Pandas plays a central role in EDA through functions like describe(), info(), value_counts(), groupby(), correlation matrices, and integration with visualization libraries like Matplotlib and Seaborn.

**4. Explain the use of the following functions:**

`a. isnull()`

Returns a boolean DataFrame/Series indicating where values are missing (NaN or None).
Used to detect missing data.

`b. notnull()`

Opposite of isnull(). Returns True for non-missing values.
Helpful to filter out rows/columns that have valid data.

`c. dropna()`

Removes rows or columns containing missing values.
Parameters: axis=0 (rows), axis=1 (columns), how='any' or 'all', thresh, inplace.

`d. fillna()`

Fills missing values with a specified value, method, or statistical measure (mean, median, mode).
Common strategies: forward fill (ffill), backward fill (bfill), or constant value.

These four functions are fundamental for missing value treatment during data wrangling.

**5. What is column transformation and why is it important in data analysis?**

Column Transformation refers to the process of creating new columns or modifying existing columns based on certain logic, calculations, or functions.

**Common ways to transform columns in Pandas:**


*   Using arithmetic operations `(df['new_col'] = df['col1'] + df['col2'])`
*   Applying functions `(df['new_col'] = df['col'].apply(lambda x: ...))`


*   Using `map()`, `replace()`, or `astype()`
*   Feature engineering (e.g., extracting day/month from date)

**Why it is important:**

Helps create more meaningful and informative features.
Improves model performance through better feature engineering.
Converts raw data into domain-specific insights.
Enables data normalization and standardization.
Supports creation of derived metrics (e.g., Total Price = Quantity × Unit Price).

Column transformation turns raw data into actionable intelligence.

**3. Sort by Energy**

**4. Rename column**

```python
df.rename(columns={'Energy (kCal)': 'Energy'}, inplace=True)
df.head()
```

**5. Summary statistics**

```python
df.describe()
```

**6. Filter rows (High Energy + High Protein)**

```python
df[(df['Energy'] > 300) & (df['Protein (g)'] > 10)]
```

**7. Group by Category (mean Energy)**

```python
df.groupby('Category')['Energy'].mean()
```

**8. Merge two DataFrames**

```python
df2 = df[['Product', 'Energy']]

merged_df = pd.merge(df, df2, on='Product')
merged_df.head()
```

**9. Handle missing values (IMPORTANT FIX)**

```python
# Since Sodium has missing values, handle it properly:
df['Sodium (mg)'] = df['Sodium (mg)'].fillna(df['Sodium (mg)'].mean())
```

```python
df.isnull().sum()
```

```python
# Drop a column (example: Fiber)
df.drop(columns=['Fiber (g)'], inplace=True)
```

**10. Apply function**

```python
df['High_Calorie'] = df['Energy'].apply(lambda x: 'Yes' if x > 300 else 'No')
df.head()
```
