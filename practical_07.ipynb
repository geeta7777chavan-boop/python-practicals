# **Theory:**

**Introduction**

Probability distributions are fundamental concepts in statistics and data science. They help us understand how data behaves and allow us to model real-world phenomena mathematically. In this assignment, we explore three important distributions: Binomial, Poisson, and Normal distributions, along with their applications and theoretical foundations.

**1. Difference between Binomial, Poisson, and Normal Distributions**

These three distributions are categorized by the type of data they model (discrete vs. continuous) and the nature of the events.

|Feature|Binomial|Poisson|Normal|
|---|---|---|---|
|Type of Data|Discrete|Discrete|Continuous|
|Key Parameters|n \(trials\), p \(prob\)|λ \(lambda/rate\)|μ \(mean\), σ \(std dev\)|
|Main Use Case|Binary outcomes|Events in time/space|Natural variations|
|Shape|Skewed \(unless p=0\.5\)|Right-skewed|Symmetric \(Bell Curve\)|
|Example Scenario|Coin flips, Pass/Fail|Network traffic, Arrivals|Heights, Error terms|

**2. Real-World Significance of the Poisson Distribution**

The Poisson distribution is vital for modeling random events that occur independently at a constant average rate. Its significance lies in resource management and "queueing theory."

1. Network Traffic: Estimating the number of data packets arriving at a router to prevent congestion.

2. Healthcare: Modeling the number of patients arriving at an emergency room to determine staffing needs.

3. Reliability Engineering: Predicting the number of failures in a machine over a specific period.

4. E-commerce: Analyzing the frequency of customer visits to a website to optimize server load.

**3. Central Limit Theorem (CLT) and the Normal Distribution**

The **Central Limit Theorem** is the bridge between non-normal data and the Normal distribution. It states that:

As the sample size ($n$) increases, the distribution of the sample means will approach a Normal distribution, regardless of the shape of the original population distribution.

**Relation to Normal Distribution:**

The CLT explains why the Normal distribution appears so often in nature. When many independent random variables are added together (like the various genetic and environmental factors that determine height), their sum tends toward a bell curve. This allows statisticians to make inferences about a population even when the underlying data is skewed or non-normal.

**4. Why is Normal Distribution widely used in ML and Statistics?**

The Normal distribution is the "gold standard" in Machine Learning for several reasons:

1. Mathematical Simplicity: Its properties (like the 68-95-99.7 rule) make it easy to calculate probabilities and intervals.

2. Algorithm Assumptions: Many algorithms, such as Linear Regression, Logistic Regression, and Gaussian Naive Bayes, assume that the underlying data or the errors (residuals) are normally distributed.

3. Optimization: In Gradient Descent, having normally distributed (and scaled) features often leads to faster convergence.

4. Standardization: The process of $Z$-score normalization relies on the Normal distribution to bring different features onto the same scale.

5. **Skewness in Distributions**

Skewness measures the asymmetry of a probability distribution around its mean.

**Positive (Right) Skew**: The "tail" on the right side is longer. The mean is typically greater than the median.
Exhibiting Distributions: Poisson (especially with low $\lambda$), Exponential, and Log-Normal.

**Negative (Left) Skew**: The "tail" on the left side is longer. The mean is typically less than the median.Exhibiting Distributions: Beta distribution (depending on parameters) or datasets like age at death in developed countries.

**Zero Skew**: The distribution is perfectly symmetrical.Exhibiting Distributions: Normal distribution.

# **Implementation**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# Binomial Distribution
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
y_binom = binom.pmf(x_binom, n, p)

# Poisson Distribution
lam = 5
x_poisson = np.arange(0, 15)
y_poisson = poisson.pmf(x_poisson, lam)

# Normal Distribution
mu, sigma = 0, 1
x_norm = np.linspace(-4, 4, 100)
y_norm = norm.pdf(x_norm, mu, sigma)

# Plotting
plt.figure()

plt.plot(x_binom, y_binom, marker='o', label="Binomial")
plt.plot(x_poisson, y_poisson, marker='x', label="Poisson")
plt.plot(x_norm, y_norm, label="Normal")

plt.title("Probability Distributions")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.legend()

plt.show()
```
