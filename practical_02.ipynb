**1. What is data ingestion and why is it important in data science? Which Python library is commonly used for data ingestion and cleaning?**

Data Ingestion is the process of acquiring, collecting, and importing raw data from various external sources (such as CSV files, databases, APIs, web scraping, cloud storage, etc.) into a system or tool where it can be stored and processed for further analysis.

**Importance in Data Science:**

1. It is the first and foundational step of any data science pipeline.
Without proper ingestion, data remains inaccessible for analysis or modeling.
2. It ensures data is brought into a structured format (e.g., DataFrame) so that cleaning, exploration, and modeling can begin.
3. Poor ingestion can lead to incomplete, corrupted, or inconsistent data, which ultimately affects the accuracy and reliability of insights and machine learning models.

Commonly used Python Library:
Pandas is the most popular and widely used library for both data ingestion and data cleaning. It provides powerful functions like read_csv(), read_excel(), read_sql(), etc., along with extensive tools for data manipulation and cleaning.

**2. Why is data cleaning necessary before performing analysis or machine learning?**

Raw data collected from real-world sources is almost never perfect. It usually contains missing values, duplicates, inconsistent formats, outliers, incorrect data types, and noise.

**Reasons why data cleaning is essential:**

1. Garbage In → Garbage Out (GIGO): Machine learning models and statistical analysis are highly sensitive to poor-quality data. Even small errors can lead to misleading conclusions or poor model performance.
2. Improves accuracy and reliability of results.
3. Helps in better feature engineering and model training.
4. Prevents errors during computation (e.g., division by zero or type mismatch).
5. Saves computational resources and time by removing unnecessary or faulty records.
6. Ensures data consistency and data integrity, which is critical for making correct business or scientific decisions.

In short, data cleaning transforms raw, messy data into high-quality, analysis-ready data.

**5. How can you identify missing values in a dataset?**

Missing values (represented as NaN, None, or blank) can severely affect analysis and machine learning models. Pandas provides simple and effective methods to detect them.

Common methods to identify missing values:

```python
# 1. Check total missing values per column
df.isnull().sum()

# 2. Check percentage of missing values
(df.isnull().sum() / len(df)) * 100

# 3. Get overall summary of missing values
df.isna().sum()

# 4. Visual check (boolean DataFrame)
df.isnull().head()

# 5. Get complete information about the dataset
df.info()
```



*   isnull() and isna() are the same (both return True where value is missing).
*   After identification, missing values can be handled using dropna() or fillna().
